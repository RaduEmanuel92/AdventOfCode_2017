```ruby
require "matrix"
class Matrix
  def []=(i, j, x)
    @rows[i][j] = x
  end
end

  
def neib_sum(m,i,j)
	sum = 	m[i-1][j-1].to_i +
	m[i][j-1].to_i 	+
	m[i+1][j-1].to_i +
	m[i+1][j].to_i 	+
	m[i+1][j+1].to_i +
	m[i][j+1].to_i +
	m[i-1][j+1].to_i +
	m[i-1][j].to_i
	return sum
end

def spiral(n)
  spiral = Array.new(2*n) {Array.new(n, nil)}     # n x n array of nils
  runs = 0.upto(n).each_cons(2).to_a.flatten  # n==5; [5,4,4,3,3,2,2,1,1,0]
  delta = [[1,0], [0,1], [-1,0], [0,-1]].cycle
  x, y, value = n/4, n/4, 1
  spiral[n/2][n/2] = 1
  for run in runs
    dx, dy = delta.next
    #print x+dx," ", y+dy
    #puts "\n"
    run.times { spiral[x+=dx][y+=dy] = (value+=1) }
  end
  spiral
end

def spiral2(n)
  spiral = Array.new(n) {Array.new(n, 0)}     # n x n array of nils
  runs = 0.upto(n).each_cons(2).to_a.flatten  # n==5; [5,4,4,3,3,2,2,1,1,0]
  delta = [[1,0], [0,1], [-1,0], [0,-1]].cycle
  x, y, value = n/2, n/2, 1
  spiral[n/2][n/2] = 1
  for run in runs
    dx, dy = delta.next
    #print spiral[2][2]
    #puts "\n"
    run.times { 
      if ((x == n/2) && (y == n/2)) 
        spiral[x+=dx][y+=dy] =  1
      else
        value = neib_sum(spiral,x,y)
        #printf("value %s\n", value)
        spiral[x+=dx][y+=dy] = value
      end 
      } 
  end
  spiral
end
 
def print_matrix(m)
  width = m.flatten.map{|x| x.to_s.size}.max
  m.each {|row| puts row.map {|x| "%#{width}s " % x}.join}
end
m = spiral(10)

m1 = spiral2(10)

print_matrix m 
print_matrix m1
```
